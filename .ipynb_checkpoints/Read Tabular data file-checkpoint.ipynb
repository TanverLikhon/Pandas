# Why do some pandas commands end with parentheses (and others don't)?


```python
movies=pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv')
```

```python
movies.head()
```

```python
movies.describe()#statistics of all numeric columns
```

```python
movies.shape
```

```python
movies.dtypes
```

```python
type(movies)
```

```python
movies.describe(include=['object'])#only describe the column which type is object
```

```python
movies.describe(include=['float64'])#only describe the column which type is float
#if we press shift+tab inside parenthesis there will show the arguments :)
```

```python
movies.describe(include=['int64'])#only describe the column which type is object
```
