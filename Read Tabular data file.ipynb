# How do I filter rows of a pandas DataFrame by column value?


```python
movies.head()
movies.shape
```

```python
#let's pretend we want to watch movies that are really atleast 200 min long
booleans =[]
for length in movies.duration:
    if length>=200:
        booleans.append(True)
    else :
        booleans.append(False)
        
#instead of for loops is_long=movies.duration>=200  can be used
        
```

```python
booleans[0:5]
```

```python
len(booleans)
```

```python
#converting booleans list to pandas series
is_long=pd.Series(booleans)
```

```python
is_long.head()
```

```python
movies[is_long]#the rows which are true in the is_long list
```

```python
movies[movies.duration>=200]
```

```python
movies[movies.duration>=200]['genre']#ormovies[movies.duration>=200].genre
```

```python
movies.loc[movies.duration>=200,'genre']#genre is a column name
```
